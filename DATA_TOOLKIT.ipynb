# Q8. Create a line plot with multiple lines using Matplotlib.

```python
x = [1, 2, 3, 4, 5]
y1 = [3, 1, 4, 6, 2]
y2 = [6, 4, 5, 8, 3]
y3 = [2, 7, 1, 4, 9]

plt.plot(x, y1 , color = 'red', marker = 'o', linestyle = '--', linewidth = 3, markersize = 6, label = 'line y1 as Tata motors')
plt.plot(x, y2 , color = 'blue', marker = 'o', linestyle = ':', linewidth = 3, markersize = 6, label = 'line y2 as Tata power')
plt.plot(x, y3 , color = 'black', marker = 'o', linestyle = '-', linewidth = 3, markersize = 6, label = 'line y3 as Tata consultancy')

plt.title('Multiple lines in one plot')
plt.xlabel('Days')
plt.ylabel('Stock price')
plt.legend()
plt.show()
```
